```python
import pandas as pd
import datetime

# Đường dẫn file của ngày hôm qua
yesterday_formatted = (datetime.datetime.now() - datetime.timedelta(days=1)).strftime("%d-%m-%Y")
yesterday_file = f"C:/Users/ADMIN/Desktop/app/data/{yesterday_formatted}_product.txt"

# Đọc file của ngày hôm qua vào DataFrame
df_yesterday = pd.read_csv(yesterday_file)

# Lấy giá trị lớn nhất của product_id từ file của ngày hôm qua
max_product_id_yesterday = df_yesterday['product_id'].max()

# Đọc file của ngày hiện tại vào DataFrame
current_date = datetime.datetime.now().strftime("%d-%m-%Y")
filename_input = f"C:/Users/ADMIN/Desktop/app/Rawdata/{current_date}_product.txt"
item_new = pd.read_csv(filename_input)

# Chỉnh sửa cột discount và price
item_new.loc[item_new['discount'] == '%', 'discount'] = '0%'
item_new['price'] = item_new['price'].str.replace('.', '').str.replace('₫', '')

# Loại bỏ các hàng trùng lặp dựa trên cột 'link'
links_in_yesterday = df_yesterday['link']
item_new = item_new[~item_new['link'].isin(links_in_yesterday)]

# Đặt lại chỉ số product_id bắt đầu từ max_product_id_yesterday + 1
item_new = item_new.reset_index(drop=True)
item_new['product_id'] = range(max_product_id_yesterday + 1, max_product_id_yesterday + 1 + len(item_new))

# Lưu DataFrame đã loại bỏ trùng lặp và cập nhật chỉ số vào file mới
filename_output = f"C:/Users/ADMIN/Desktop/app/data/{current_date}_product_no_duplicates.txt"
item_new.to_csv(filename_output, sep=',', index=False)
```

```python
file_name = r'C:\Users\ADMIN\Desktop\app\Rawdata\31-01-2024_prodcut_detail.txt'

with open(file_name, 'r', encoding='utf-8') as file:
    for line in file:
        print(line)
```

```python
lines = []
with open(file_name, 'r', encoding='utf-8') as file:
    for line in file:
        if 'http' in line:
            lines.append(line)
```

```python
print(len(lines))
```

```python
# Danh sách các từ khóa để tìm kiếm thông tin, bao gồm cả link
keywords = ['Mã Giảm Giá','Thương hiệu', 'Xuất xứ thương hiệu', 'Xuất xứ', 'Hạn sử dụng', 'được bảo hành không','Thời gian bảo hành','link']

# Khởi tạo DataFrame để lưu thông tin
df = pd.DataFrame(columns=keywords)

# Duyệt qua mỗi dòng giả lập và trích xuất thông tin
for line in lines:
    # Tách chuỗi dựa trên dấu phẩy và dấu chấm
    parts = [part.strip() for part in line.split(',') if part.strip()]

    # Tạo từ điển để lưu thông tin cho dòng hiện tại
    product_info = {key: None for key in keywords}

    # Xác định và lưu thông tin vào từ điển
    for i, part in enumerate(parts):
        # Đối với link, tìm chuỗi bắt đầu bằng 'http' và lưu vào từ điển
        if 'http' in part:
            product_info['link'] = part
        if 'Mã Giảm Giá' in part:
            product_info['Mã Giảm Giá'] = part[0]
        else:
            for key in keywords:
                if key in part and key != 'link' and key != 'Mã Giảm Giá':
                    try:
                        # Chú ý: giả định rằng thông tin luôn nằm sau từ khóa
                        product_info[key] = parts[i + 1]
                    except IndexError:
                        product_info[key] = None

    # Thêm thông tin vào DataFrame
    df.loc[len(df)] = product_info

    # df = df.append(product_info, ignore_index=True)

# Thay thế các giá trị None bằng NaN trong DataFrame
df.fillna(value=pd.NA, inplace=True)
```

```python
df['product_id'] = df['link'].map(link_to_product_id)
```

```python
df = df.drop('link', axis=1)
```

```python
df.to_csv('data/product_info.csv', index=False)
```

```python
# Vì không thể trực tiếp đọc file "my_file.txt" từ hệ thống của bạn, chúng ta sẽ mô phỏng quá trình này.
# Bạn sẽ cần thay 'my_file.txt' bằng đường dẫn chính xác đến file trên hệ thống của bạn khi chạy mã này.

# Mô phỏng nội dung file với một số dòng có chứa 'http' và một số không
file_content = """Sản phẩm 1, Thông tin, http://example.com/product1\n
                  Sản phẩm 2, Thông tin, No link here\n
                  Sản phẩm 3, Thông tin, https://example.com/product3\n
                  Sản phẩm 4, Thông tin, Just text"""

# Giả lập việc đọc file bằng cách tách nội dung giả lập thành từng dòng
lines = file_content.split('\n')

# Lọc ra những dòng có chứa 'http'
lines_with_http = [line for line in lines if 'http' in line]

# In độ dài ban đầu và sau khi lọc
print(f"Length before filtering: {len(lines)}")
print(f"Length after filtering: {len(lines_with_http)}")

# In ra nội dung sau khi lọc
print((lines_with_http))
```
